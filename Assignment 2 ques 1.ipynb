# predicting employee exit data

```python
# import standard libraries
```

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# load data
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/tranghth-lux/data-science-complete-tutorial/master/Data/HR_comma_sep.csv.txt")
df
```

```python
# the dependent column is "left"

df.left.unique()
# since the dependent variable is categorical in nature for prediction we can use knn_classification logat=rithm for prediction
```

# Data preprocessing

```python
# 1) missing value check
```

```python
df.isnull().sum()  # no missing values present in the data
```

```python
#2) Feature encoding 

 # we have to convert sales and salary column to numerical

 # sales column is nominal variable (use one hot encoder)

# and salary column is ordinal variable  (use ordinal encoding)
```

```python
# for one hot encoding use dummies

df_ohe=pd.get_dummies(df["sales"])
df_ohe
```

```python
#combine both datas
```

```python
df=pd.concat([df,df_ohe],axis=1)
df
```

```python
# drop sales column
```

```python
df=df.drop(["sales"],axis=1)
df
```

```python
#for salary column we need to use the ordinal encoder
```

```python
from sklearn.preprocessing import OrdinalEncoder
od=OrdinalEncoder()
od
```

```python
df["salary"]=od.fit_transform(df[["salary"]])
df
```

```python
df.describe()
```

# feature scaling

```python
# we have to scale  "average_montly_hours" column into a standard format
```

```python
from sklearn.preprocessing import RobustScaler
rb=RobustScaler()
rb
```

```python
df["average_montly_hours"]=rb.fit_transform(df[["average_montly_hours"]])
df
```

```python
#preprocessing is completed on the dataset
```

# knn neighbour

```python
#divding data to x and y
```

```python
x=df.drop(["left"],axis=1)
y=df["left"]
```

```python
#split into train and test
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#importing and fiting knn neighbour
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=123)
knn
```

```python
# for finding k value

import math
math.sqrt(14999) # so we can choose 123 as k values
```

```python
knn.fit(x_train,y_train)
```

```python
#prediction
```

```python
y_pred=knn.predict(x_test)
y_pred
pred_value=list(y_pred)
for i in pred_value:
    print(i,end="")
```

```python
#evaluation
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#the accuracy score is equal to 93%
```
